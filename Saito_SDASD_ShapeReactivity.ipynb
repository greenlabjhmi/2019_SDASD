```python
filenames = {}
with open('/home/kazuki/^data_analysis/^SHAPE_analysis/Mustoe2018_data/transcripts.txt','r') as tsf:
    for line in tsf:
        if line[:10]=='transcript':
            continue
        linelist = line.split(',')
        filename = linelist[2]+'-'+linelist[3]+'.shape'
        filenames[linelist[0]]=filename
```

```python
parentpath = '/home/kazuki/^data_analysis/^SHAPE_analysis/Mustoe2018_data/incell_SHAPE/'

Gene_ShapeFile_positions= {}
with open('/home/kazuki/^data_analysis/^SHAPE_analysis/Mustoe2018_data/genes.txt','r') as gf:
    for line in gf:
        if line[:4]=='gene':
            continue
        linelist = line.split(',')
        gene = linelist[0]
        transcript = linelist[5][:-1]
        Tstart = linelist[3]
        unfold_start = str(int(Tstart)-25)
        unfold_end = str(int(Tstart)+25)
        parent_ct = parentpath+filenames[transcript]
        Gene_ShapeFile_positions[gene] = [parent_ct,unfold_start,unfold_end]
```

```python
#import statistics
import math
import numpy as np

ShapeReactivity = {}
for key in Gene_ShapeFile_positions.keys():
    ReactivityRange = [str(x) for x in range(int(Gene_ShapeFile_positions[key][1]),int(Gene_ShapeFile_positions[key][2]))]
    ShapeList = []
    SeqList = []
    with open(Gene_ShapeFile_positions[key][0]) as shapefile:
        for line in shapefile:
            linelist = line.split()
            if linelist[0] in ReactivityRange:
                ShapeList.append(float(linelist[1]))
                SeqList.append(linelist[3])
     
    Seq = ''.join(SeqList)
    x = np.array(ShapeList)
    ShapeReactivity[key] = []
    if len(x)==0 or float(len(x[~np.isnan(x)]))/len(x) < 0.8:
        ShapeReactivity[key].append(np.nan)
        ShapeReactivity[key].append(np.nan) 
    else:
        median = np.median(x[~np.isnan(x)])
        ShapeReactivity[key].append(median)
        ShapeReactivity[key].append(Seq)   
```

```python
import pickle
with open('ShapeReactivity.pickle', 'wb') as pickle_f:
    pickle.dump(ShapeReactivity,pickle_f)
```
